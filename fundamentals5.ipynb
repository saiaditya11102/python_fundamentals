```python
# prompt: create same in pie chart

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Universities.csv')

# Assuming 'SAT' column represents the values for the pie chart
# and you want to group them into categories for a more meaningful pie chart
# For example grouping by ranges of SAT scores

# Define bins (example ranges for SAT scores)
bins = [0, 1000, 1200, 1400, 1600, 2000]  # Adjust these bins as needed
labels = ['<1000', '1000-1199', '1200-1399', '1400-1599', '1600+'] #labels for each bin
df['SAT_Group'] = pd.cut(df['SAT'], bins=bins, labels=labels, right=False)

# Calculate the counts for each SAT group
sat_counts = df['SAT_Group'].value_counts()


plt.figure(figsize=(8, 6))
plt.pie(sat_counts, labels=sat_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of SAT Scores')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
